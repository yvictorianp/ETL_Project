```python
import pandas as pd
from sqlalchemy import create_engine
pd.set_option('display.max_rows', None)
```

```python
financial_data_file = "../Resources/2018_Financial_Data 2.csv"
return_data_file = "../Resources/Return.csv"
```

```python
financial_data_df = pd.read_csv(financial_data_file)
financial_data_df.head()
```

```python
return_data_file_df = pd.read_csv(return_data_file)
return_data_file_df.head()
```

```python
financial_data_df.columns.tolist()
```

```python
#Unnamed: 0,Revenue,Net Income, EPS,Dividend per Share, returnOnEquity, debtRatio, Market Cap
reduced_financial_data = financial_data_df [["Unnamed: 0", "Revenue", "Net Income", "EPS", "Dividend per Share",
                                            "returnOnEquity", "debtRatio", "Market Cap"]]
reduced_financial_data.head()
```

```python
reduced_financial_data = reduced_financial_data.rename (columns={'Unnamed: 0':'ticker','Revenue': 'revenue',
                                                                 'Net Income':'net_income','EPS':'eps',
                                                                 'Dividend per Share':'dividend_per_share',
                                                                 'returnOnEquity': 'roe','debtRatio': 'debt_ratio',
                                                                'Market Cap':'market_cap'})
reduced_financial_data.head()
```

```python
renamed_return_data_df = return_data_file_df.rename (columns={'TICKER': 'ticker', 'NAME':'company_name', 'Date_12/29/2017': 
                                                              'stock_price_as_of_12292017', 'Date_12/31/2018':
                                                              'stock_price_as_of_12312018', 'Return':'return'})
renamed_return_data_df.head()
```

```python
reduced_financial_data.columns
```

```python
reduced_financial_data = reduced_financial_data.sort_values(by='revenue', ascending = False)
reduced_financial_data.head()
```

```python
reduced_financial_data_df= reduced_financial_data.reset_index(drop=True)
reduced_financial_data_df.head()
```

```python
clean_financial_data = reduced_financial_data_df[0:101]
clean_financial_data.head()
```

```python
combined_financial_return_data = pd.merge(clean_financial_data, renamed_return_data_df, on="ticker", how="outer")
combined_financial_return_data.head()
```

```python
#topUSCompanies
final_financial_return_data = pd.merge(clean_financial_data, renamed_return_data_df, on="ticker", how="inner")
final_financial_return_data.head()
final_financial_return_data.dtypes
```

```python
#numberformatting
final_financial_return_data ["revenue"] = final_financial_return_data ["revenue"].map("${:,.0f}".format)
final_financial_return_data ["net_income"] = final_financial_return_data ["net_income"].map("{:,.0f}".format)
final_financial_return_data ["eps"] = final_financial_return_data ["eps"].map("{:.2f}".format)
final_financial_return_data ["dividend_per_share"] = final_financial_return_data ["dividend_per_share"].map("${:.2f}".format)
final_financial_return_data ["roe"] = final_financial_return_data ["roe"].map("{:.2f}%".format)
final_financial_return_data ["debt_ratio"] = final_financial_return_data ["debt_ratio"].map("{:.2f}%".format)
final_financial_return_data ["market_cap"] = final_financial_return_data ["market_cap"].map("${:,.0f}".format)
final_financial_return_data ["stock_price_as_of_12292017"] = final_financial_return_data ["stock_price_as_of_12292017"].map("${:.2f}".format)
final_financial_return_data ["stock_price_as_of_12312018"] = final_financial_return_data ["stock_price_as_of_12312018"].map("${:.2f}".format)
final_financial_return_data.head()
final_financial_return_data.dtypes

```

```python
#connect to database
rds_connection_string = "postgres:postgres@localhost:5432/financial_data_db"
#username & password may be postgres
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
#check for table
engine.table_names()
```

```python
#use pandas to load csv
final_financial_return_data.to_sql(name='financial_data', con=engine, if_exists='append', index=False)
```

```python
#confirming data is in query
pd.read_sql_query('select * from financial_data', con=engine).head()
```
